```python
# Nguyễn Minh Đức - ITITIU21045_TMCLab4_Q2c
import numpy as np
A = np.array([
  [1, 1, -1, -3],
  [6, 2, 2, 2],
  [-3, 4, 1, 1]
], dtype=float)
n = len(A)

for i in range(n):
  pivot = A[i, i]
  A[i, :] /= pivot

  for j in range(n):
    if j != i:
      factor = A[j, i]
      A[j, :] -= factor * A[i, :]

x = A[:, -1]

for i in range(n):
  print(f'x{i+1} = {x[i]:.2f}')
```
