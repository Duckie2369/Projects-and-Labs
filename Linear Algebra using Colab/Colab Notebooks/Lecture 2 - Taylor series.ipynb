```python
xi = 0.2
xip = 1
f0 = np.exp(-xi)
f0
```

```python
# true value
f_true = np.exp(-xip)

# relative error
e0 = np.abs(f_true-f0)/f_true*100
print(e0)
```

```python
# first order approx.
f1 = f0 + (-np.exp(-xi))*(xip-xi)
print(f1)

# True relative error
e1 = np.abs(f_true-f1)/f_true*100
print(e1)
```

```python
# second oder approximation
f2 = f1+ np.exp(-0.2)/2*(1 - 0.2)**2
print(f2)

e2 = np.abs(f_true-f2)/f_true*100
```

```python
print (f0, e0)
print (f1, e1)
print (f2, e2)
```

```python
 import numpy as np
```

```python
def f(x):
  return -0.1*x**4 - 0.15*x**3 - 0.5*x**2 - 0.25*x+1.25
```

```python
fp_true = -0.9125
xi = 0.5
h = 0.25
xiplus = xi + h
ximinus = xi - h
```

```python
# backward
fp_back = (f(xi) - f(ximinus))/h
e_back = np.abs(fp_true - fp_back)
print(fp_back, e_back)
```

```python
# forward
fp_for = (f(xiplus) - f(xi))/h
e_for = np.abs(fp_true - fp_for)
print(fp_for, e_for)
```

```python
# center
fp_cen = (f(xiplus) - f(ximinus))/(2*h)
e_cen = np.abs(fp_true - fp_cen)
print(fp_cen, e_cen)
```
